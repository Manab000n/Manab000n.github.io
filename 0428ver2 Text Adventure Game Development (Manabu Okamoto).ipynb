Description:

Now you are a young, poor and miserable Italian American professional boxer named Rocky!!
You have a large number of issues based on unstable financial situation and you are not motivated.
It is the time to overcome these difficulties and change your life!! 
Become the world champion within 3 minutes, following in the footsteps of Rocky Balboa.

This game is based on the Rocky series - Stallone, S. (Writer/Director). (1976-2018). Rocky [Film]. United Artists.

```python
import random

# Function to display win message
def win():
    SUCCESS_MESSAGE = "Great job! You've become the champion of the world!! Adrian, Mickey, and your fans are impressed by your brave story"
    print(SUCCESS_MESSAGE)

# Function to display fail message
def fail():
    FAILURE_MESSAGE = "Game Over!! Sorry, you can't win and change your life."
    print(FAILURE_MESSAGE)

# Function to get valid choice with improved flexibility and error handling
def get_valid_choice(options, prompt="Enter your choice"):
    valid_responses = {str(index + 1): option for index, option in enumerate(options)}
    valid_responses.update({option.lower(): option for option in options})  # Include other acceptable forms
    
    while True:
        choice = input(f"{prompt} ({'/'.join(valid_responses.keys())}): ").lower()
        if choice in valid_responses:
            return choice
        else:
            print("Invalid choice. Please try again.")

# Randomized stage 1
def stage_1():
    LOCATIONS = random.sample(["Mickey's gym", "Paulie's place", "Adrian's home", "his meat company office", "Training Ground"], 2)
    print("You are a young and miserable boxer in Philadelphia. It is time to change your life!! (Round 1)")
    print("You have two options (HINT: Success always demands commitment):")
    
    for i, location in enumerate(LOCATIONS, start=1):
        print(f"{i}. Train hard at {location}")

    choice_stage_1 = get_valid_choice(['1', '2'], prompt="Choose your path")
    if choice_stage_1 == '1':
        print(f"You chose to train hard at {LOCATIONS[0]}. Good decision! You clearly became mature and stronger than before.")
        return True
    else:
        print(f"The person whom you stay with is the key to your success, and {LOCATIONS[1]} is not ideal for a future champion.")
        return False

# Randomized stage 2
def stage_2():
    LOCATIONS = random.sample(["Mickey's gym", "Paulie's place", "Adrian's home", "his meat company office", "Training Ground"], 2)
    print("You are given a day off from hard training by your strict coach Mickey. (Round 2)")
    print("You have two options:")
    
    for i, location in enumerate(LOCATIONS, start=1):
        print(f"{i}. Spend the day at {location}")

    choice_stage_2 = get_valid_choice(['1', '2'], prompt="Choose your path")
    if choice_stage_2 == '1':
        print("You get a girlfriend who gives you unconditional love. Now you are highly motivated to be a champion for her and your baby inside her.")
        return True
    else:
        print(f"The person whom you stay with is the key to your success. {LOCATIONS[1]} is not ideal for a future champion even though his sibling is amazing.")
        return False

# Randomized stage 3
def stage_3():
    LOCATIONS = random.sample(["Mickey's gym", "Paulie's place", "Adrian's home", "his meat company office", "Training Ground", "1st Round Knockout"], 2)
    print("You've reached the World match with current champion Apollo Creed. (Round 3)")
    print("How do you approach this challenge?(HINT: There is no magic):")
    
    for i, location in enumerate(LOCATIONS, start=1):
        print(f"{i}. Stay focused on your strategy - {location}")

    choice_stage_3 = get_valid_choice(['1', '2'], prompt="Choose your strategy")
    if choice_stage_3 == '1':
        print("Staying focused and true to your training pays off. You win the championship!")
        return True
    else:
        print("Resilience is the key to winning the hard competition. Try again.")
        return False

# Introduction and game flow
def main():
    print("Welcome to the Rocky Adventure Game!")
    print("Based on the Rocky series - Stallone, S. (Writer/Director). (1976-2018). Rocky [Film Series]. United Artists.")
    print("Your goal is to become the champion within 3 minutes!!")
    
    # Stage progression
    if stage_1():
        if stage_2():
            if stage_3():
                win()
            else:
                fail()
        else:
            fail()
    else:
        fail()

# Run the game
main()
```
